<p><h3>Predictive Question/Exploratory Analysis and Visualization</h3>
Can Average played hours per user and age of user predict the experience of the player? <p></p>

To answer this question, we can first tidy the players.csv dataframe by cleaning up column names, converting columns to appropriate data types (response variable for KNN classification must be a factor data type), and filter for desired predictors. We filter out played hours less than 0 since it is not meaningful for our analysis. Next, we group by both experience and age, and aggregate played hours to better see their relationship for each skill level. To analyze the total hours played per user in each skill level group, we can group by experience, then aggregate by obatining the sum of played hours and the number of users to compute avg_hours_per_user. This can subsequently help us assess whether these different variables are useful for predictive analysis.

```r
players <- players|>
clean_names() |>
mutate(experience = as.factor(experience))|>
filter(played_hours > 0.0)

age_hours <- players|>
group_by(experience, age)|>
summarize(avg_hours = mean(played_hours))

experience_by_hours <- players|>
group_by(experience)|>
summarize(total_hours = sum(played_hours), 
          users = n(), 
          avg_hours_per_user = total_hours/users)
experience_by_hours
```

We can now visualize the relationship between age and average played hours using the dataframe "age_hours" and the relationship between experience and total hours using the dataframe "experience_by_hours".

```r
options(repr.plot.width = 8, repr.plot.height = 8)
summary_players_plot <- age_hours |>
ggplot(aes(x = age, y = avg_hours, color = experience)) +
geom_point() +
labs(x = "Ages of Users", y = "Hours Spent on Gaming", title = "Average Played Hours by Age", color = "Skill Level") +
theme(text = element_text(size = 17))
summary_players_plot
```

From the first plot, it is notable that there are no distinct relationship between the age variable and the average hours played. The data points are dispersed across all age groups, showing no upward or downward trend between the two variables. This suggests that age of the user cannot determine the time they spent on gaming. Sorting experience by colour, we can see that neither age nor average played hours correlate with the skill level of the user as the colours are spread randomly with no distinct cluster. 

```r
options(repr.plot.width = 8, repr.plot.height = 8)
experience_hours_plot <- experience_by_hours |>
ggplot(aes(x = experience, y = avg_hours_per_user)) +
geom_bar(stat="identity") +
labs(x = "Skill Level of Users", y = "Average Hours Played per User", title = "Experience by Total Hours") +
theme(text = element_text(size = 17))
experience_hours_plot
```

The second graph depicts the average hours played from each skill level. We can see that Regulars group have the highest average hours played while the Veterans group has the lowest. This suggests that there is a non-linear relationship between average hours and experience, where more time spent on gaming does not necessarily correspond to higher skill level.


<h3>Methods and Plan</h3>
<p>To address the predictive question, the KNN classification model can be used. This method is appropriate because the chosen response variable is categorical rather than numeric and KNN does not assume the distribution of data. However, KNN assumes that users with similar total hours and ages correspond to similar skill levels since it predicts using the distance metric. As seen in the visualizations, the two variables may be weak predictors due to the lack of visible trends or clusters, meaning the KNN classification model is inherently limited in making accurate predictions using these predictors. To compare and select the model, five-fold validation sets can be used to determine the best k neighbour, measured based on the accuracy metric. To process the data, we can first split the data into training and testing sets with proportions 0.7 training and 0.3 testing to provide enough data for the model to learn and later predict on unseen data. Then, cross validation will be used to select hyperparameters. Finally, the best k will be used to retrain the model and this model will be fitted on the test set to make predictions. </p>
